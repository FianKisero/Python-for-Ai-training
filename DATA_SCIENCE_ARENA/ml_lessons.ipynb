```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare our data
# (Scikit-learn expects X to be a 2D matrix, so we use .reshape(-1, 1))
X = np.array([10, 20, 30]).reshape(-1, 1)  # Features: Epochs
y = np.array([0.85, 0.91, 0.95])           # Target: Accuracy

# 2. Instantiate the model
model = LinearRegression()

# 3. Train the model (fit the data)
model.fit(X, y)

# 4. Predict the accuracy of a run with 40 epochs!
future_epoch = np.array([[40]])
predicted_accuracy = model.predict(future_epoch)

print(f"Predicted Accuracy for 40 epochs: {predicted_accuracy[0]:.4f}")
```
